```python
#This notebook is a 'playground' to explore how to calculate solar zenith angle (SZA) adjustments to visible/near-IR imagery
#Author: Kevin Thiel
#Created: February 2026

#Used to calculate solar zenith angle (SZA) correction
from pyorbital.astronomy import cos_zen as pyob_cos_zen #Specific calculation needed for solar zenith angle
import numpy as np
import netCDF4 as nc
import pandas as pd

#Used for plotting
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from datetime import datetime
```

```python
# Source: https://www.star.nesdis.noaa.gov/atmospheric-composition-training/python_abi_lat_lon.php
# Please acknowledge the NOAA/NESDIS/STAR Aerosols and Atmospheric Composition Science Team if using any of this code in your work/research!

def calculate_degrees(file_id):
    
    # Read in GOES ABI fixed grid projection variables and constants
    x_coordinate_1d = file_id.variables['x'][:]  # E/W scanning angle in radians
    y_coordinate_1d = file_id.variables['y'][:]  # N/S elevation angle in radians
    projection_info = file_id.variables['goes_imager_projection']
    lon_origin = projection_info.longitude_of_projection_origin
    H = projection_info.perspective_point_height+projection_info.semi_major_axis
    r_eq = projection_info.semi_major_axis
    r_pol = projection_info.semi_minor_axis
    
    # Create 2D coordinate matrices from 1D coordinate vectors
    x_coordinate_2d, y_coordinate_2d = np.meshgrid(x_coordinate_1d, y_coordinate_1d)
    
    # Equations to calculate latitude and longitude
    lambda_0 = (lon_origin*np.pi)/180.0  
    a_var = np.power(np.sin(x_coordinate_2d),2.0) + (np.power(np.cos(x_coordinate_2d),2.0)*(np.power(np.cos(y_coordinate_2d),2.0)+(((r_eq*r_eq)/(r_pol*r_pol))*np.power(np.sin(y_coordinate_2d),2.0))))
    b_var = -2.0*H*np.cos(x_coordinate_2d)*np.cos(y_coordinate_2d)
    c_var = (H**2.0)-(r_eq**2.0)
    r_s = (-1.0*b_var - np.sqrt((b_var**2)-(4.0*a_var*c_var)))/(2.0*a_var)
    s_x = r_s*np.cos(x_coordinate_2d)*np.cos(y_coordinate_2d)
    s_y = - r_s*np.sin(x_coordinate_2d)
    s_z = r_s*np.cos(x_coordinate_2d)*np.sin(y_coordinate_2d)
    
    # Ignore numpy errors for sqrt of negative number; occurs for GOES-16 ABI CONUS sector data
    np.seterr(all='ignore')
    
    abi_lat = (180.0/np.pi)*(np.arctan(((r_eq*r_eq)/(r_pol*r_pol))*((s_z/np.sqrt(((H-s_x)*(H-s_x))+(s_y*s_y))))))
    abi_lon = (lambda_0 - np.arctan(s_y/(H-s_x)))*(180.0/np.pi)
    
    return abi_lat, abi_lon, x_coordinate_2d, y_coordinate_2d
```

```python
#Reading in ABI netCDF file (GOES-East (GOES-19) ABI CONUS, Channel 2 imagery
dset = nc.Dataset('ABI19-data/OR_ABI-L2-CMIPC-M6C02_G19_s20260401431177_e20260401433550_c20260401434080.nc','r')
cmi = dset['CMI'][:] #Radiances from channel 2

#Calculating the lats/lons from the geostationary projection coordinates (radians)
lats, lons, x_coordinate_2d, y_coordinate_2d = calculate_degrees(dset)
```

```python
#Getting the time to use in solar zenith angle calculation
time = pd.to_datetime(dset.time_coverage_start)

#Calculating the cosine of solar zenith angle
cos_zen_grid = pyob_cos_zen(time, lons, lats)

#Masking out pixels with a sun angle < 0 and too close to sunrise/sunset
#The 0.02 value can be changed to remove large sza adjustments
cos_zen_grid[cos_zen_grid<0.02] = 1

#Applying the solar zenith angle adjustment to the ABI reflectance factor data
cza = cmi / cos_zen_grid
```

# Plotting without/with the SZA adjustment

```python
fig = plt.figure(figsize=(16, 9))
plt.imshow(np.sqrt(cmi), cmap='grey',vmin=0, vmax=1.3)
plt.axis('off')
plt.title('GOES-19 CONUS ABI, Ch. 2 Reflectance Factor - No Adjustment'+'  '+str(time), fontsize='large')
plt.colorbar()
plt.savefig('test-figs/ABI19-CONUS-CH2-RF-rawsqrt-dpi300.png', dpi=300, facecolor='w', edgecolor='w')
```

```python
fig = plt.figure(figsize=(16, 9))
plt.imshow(np.sqrt(cza), cmap='binary_r',vmin=0, vmax=1.3)
plt.axis('off')
plt.title('GOES-19 CONUS ABI, Ch. 2 Reflectance Factor - SZA Adjustment'+'  '+str(time), fontsize='large')
plt.colorbar()
plt.savefig('test-figs/ABI19-CONUS-CH2-RF-SZAsqrt-dpi300.png', dpi=300, facecolor='w', edgecolor='w')
```

# Plotting for a hypothetical CONUS scene to see what the SZA adjustments would look like at different times

```python
geo_crs = ccrs.Geostationary(central_longitude=dset.variables['goes_imager_projection'].longitude_of_projection_origin, satellite_height=dset.variables['goes_imager_projection'].perspective_point_height, sweep_axis='x')
extent = np.concat((dset.variables['x_image_bounds'][:],dset.variables['y_image_bounds'][:]), axis=0)

plot_sza = pyob_cos_zen(datetime(2026,2,9,14,33), lons, lats)

fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(1, 1, 1, projection=geo_crs)

a = ax.imshow(plot_sza, cmap='gray', extent=extent, transform=geo_crs, vmin=-1, vmax=1, zorder=1)
b = ax.contour(x_coordinate_2d, y_coordinate_2d, plot_sza[::-1,:], levels=np.arange(-1.,1.05,.1), transform=geo_crs, zorder=2)
ax.clabel(b, fontsize=10)
plt.colorbar(a)
plt.colorbar(b)
```

# Testing how SZA changes for a single point (lat/lon) with time

```python
t_list = pd.date_range(start=datetime(2026,6,4,12,0), end=datetime(2026,6,5,12,0), freq='5min')
lat_list = np.ones(len(t_list)) * 39
lon_list = np.ones(len(t_list)) * -82

cos_zen_list = np.array([pyob_cos_zen(t_list[i], lon_list[i], lat_list[i]) for i in range(len(t_list))])

cos_zen_list[cos_zen_list<0] = 0
```

```python
plt.plot(t_list, cos_zen_list)
plt.grid(True)
plt.xticks(rotation=45)
plt.ylim(0,4)
```

```python
plt.plot(t_list, 1/cos_zen_list)
plt.grid(True)
plt.xticks(rotation=45)
plt.ylim(0,4)
```
