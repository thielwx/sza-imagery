```python
#Creating RGB imagery from satellite bands
import numpy as np
import netCDF4 as nc
import pandas as pd
import sza_calc as sza #local python script with functions

#Used for plotting
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from datetime import datetime
```

```python
#Loading the datsets
g_dset = nc.Dataset('../test/20260224-1411-combo/OR_ABI-L2-CMIPC-M6C02_G19_s20260551411177_e20260551413550_c20260551414061.nc', 'r')
b_dset = nc.Dataset('../test/20260224-1411-combo/OR_ABI-L2-CMIPC-M6C05_G19_s20260551411177_e20260551413551_c20260551414043.nc', 'r')
r_dset = nc.Dataset('../test/20260224-1411-combo/OR_ABI-L2-CMIPC-M6C13_G19_s20260551411177_e20260551413563_c20260551414061.nc', 'r')
```

```python
t_start = datetime.now()
dcpd_rgb_sza = sza.rgb_creator_dcpd(r_dset, g_dset, b_dset, scene='CONUS', sza_adjustment=True)
print (datetime.now() - t_start)
```

```python
dcpd_rgb = sza.rgb_creator_dcpd(r_dset, g_dset, b_dset, scene='CONUS', sza_adjustment=False)
```

```python
def plot_data(dset):
    geo_crs = ccrs.Geostationary(central_longitude=dset.variables['goes_imager_projection'].longitude_of_projection_origin, satellite_height=dset.variables['goes_imager_projection'].perspective_point_height, sweep_axis='x')
    extent = np.concat((dset.variables['x_image_bounds'][:],dset.variables['y_image_bounds'][::-1]), axis=0) * dset.variables['goes_imager_projection'].perspective_point_height
    return geo_crs, extent
```

```python
extent
```

```python
pd.to_datetime(r_dset.time_coverage_start).strftime('%Y-%m-%d %H%M UTC')
```

```python
geo_crs, extent = plot_data(g_dset)
```

```python
fig = plt.figure(figsize=(14,9))
ax = fig.add_subplot(1, 1, 1, projection=geo_crs)
a = ax.imshow(dcpd_rgb, extent=extent, transform=geo_crs)
ax.coastlines(resolution='50m', color='black', linewidth=1)
ax.add_feature(ccrs.cartopy.feature.STATES)
ax.set_extent([-108 ,-95, 35, 43])
ax.set_title('Day Cloud Phase Distinction RGB (no SZA) - '+str(r_dset.time_coverage_start))
plt.savefig('../../../../Desktop/sza-playground/test-figs/dcpd-rgb-raw-cplains-dpi300.png', dpi=300)
```

```python
fig = plt.figure(figsize=(14,9))
ax = fig.add_subplot(1, 1, 1, projection=geo_crs)
ax.imshow(dcpd_rgb_sza, extent=extent, transform=geo_crs)
ax.coastlines(resolution='10m', color='k', linewidth=0.5)
ax.add_feature(cfeature.STATES, linewidth=0.5, edgecolor='k')
ax.set_extent([-108 ,-95, 35, 43])
ax.set_title('Day Cloud Phase Distinction RGB (with SZA) - '+str(r_dset.time_coverage_start))
plt.savefig('../../../../Desktop/sza-playground/test-figs/dcpd-rgb-sza-cplains-dpi300.png', dpi=300)
```
